# What is Markov Decision Process?

A Markov Decision Process (MDP) is a mathematical framework used to solve sequential decision-making problems.

An MDP consists of five components:

1. States (S)
2. Actions (A)
3. Transition Probability (P)
4. Reward Function (R)
5. Discount Factor (γ)

```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
states=[]

for i in range(rows):
    for j in range(cols):
        states.append((i,j))

states
```

```python
actions=["UP","DOWN","LEFT","RIGHT"]

print(actions)
```

```python
reward=np.full((rows,cols),-1)

reward[3][3]=10

reward
```

```python
plt.figure(figsize=(5,5))

plt.imshow(reward,cmap='viridis')

plt.colorbar()

plt.title("Reward Matrix")

plt.show()
```

```python
current_state=(0,0)

print("Initial State :",current_state)
```

```python
def take_action(state,action):

    row,col=state

    if action=="UP":
        row=max(row-1,0)

    elif action=="DOWN":
        row=min(row+1,rows-1)

    elif action=="LEFT":
        col=max(col-1,0)

    elif action=="RIGHT":
        col=min(col+1,cols-1)

    return(row,col)
```

```python
state=(0,0)

new_state=take_action(state,"RIGHT")

print("Old State :",state)

print("New State :",new_state)
```

```python
state=(0,0)

for i in range(10):

    action=random.choice(actions)

    state=take_action(state,action)

    print(action,"---->",state)
```

```python
path=np.zeros((rows,cols))

state=(0,0)

path[state]=1

for i in range(20):

    action=random.choice(actions)

    state=take_action(state,action)

    path[state]+=1
```

```python
plt.figure(figsize=(5,5))

plt.imshow(path,cmap="Reds")

plt.colorbar()

plt.title("Visited States")

plt.show()
```
